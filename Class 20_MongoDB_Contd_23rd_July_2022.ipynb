**SQL:** Structured Query Language: Creating Schema before inserting data. We scale vertically.

**NoSQL:** Not only SQL: No need to create schema before creating table: Document based data (json based dataset/dictionary format) can be stored in MongoDB. Not supposed to create complex data modelling here unlike SQL database. We can try scale horizontally.
Cassandra, Neo4j, Elastic, Hbase are other examples of NoSQL Databases apart from MongoDB. Indexing and Searching becomes easy. We always ty to think about complexity, use-case etc while selecting different database.

**Django** is a python framework that connect frontend and backend. It is not a database.

**MongoDB Atlas:** Cloud Hosted MongoDB service provided by AWS, Azure, Google: Flexibility present to choose different storage options and features according to our requirements. M10 or higher version will be able to connect with BI Tool such as PowerBI

**MongoDB Compass:** Offline Version of MongoDB Service

Creating Collection/Document inside Database in NoSQL system is same as creating table inside Databse in SQL system. For every document, MongoDB will give object ID which will help us in index/search operation. We can even give our own ID of each document.

```python
#test3.py
import pymongo
client = pymongo.MongoClient("mongodb+srv://gauravratan:ratan94@cluster0.hunum.mongodb.net/?retryWrites=true&w=majority") #Connectivity between python and MongoDB
db = client.test

data =  [
        {
            "item": "canvas",
            "qty": 100,
            "size": {"h": 28, "w": 35.5, "uom": "cm"},
            "status": "A",
        },
        {
            "item": "journal",
            "qty": 25,
            "size": {"h": 14, "w": 21, "uom": "cm"},
            "status": "A",
        },
        {
            "item": "mat",
            "qty": 85,
            "size": {"h": 27.9, "w": 35.5, "uom": "cm"},
            "status": "A",
        },
        {
            "item": "mousepad",
            "qty": 25,
            "size": {"h": 19, "w": 22.85, "uom": "cm"},
            "status": "P",
        },
        {
            "item": "notebook",
            "qty": 50,
            "size": {"h": 8.5, "w": 11, "uom": "in"},
            "status": "P",
        },
        {
            "item": "paper",
            "qty": 100,
            "size": {"h": 8.5, "w": 11, "uom": "in"},
            "status": "D",
        },
        {
            "item": "planner",
            "qty": 75,
            "size": {"h": 22.85, "w": 30, "uom": "cm"},
            "status": "D",
        },
        {
            "item": "postcard",
            "qty": 45,
            "size": {"h": 10, "w": 15.25, "uom": "cm"},
            "status": "A",
        },
        {
            "item": "sketchbook",
            "qty": 80,
            "size": {"h": 14, "w": 21, "uom": "cm"},
            "status": "A",
        },
        {
            "item": "sketch pad",
            "qty": 95,
            "size": {"h": 22.85, "w": 30.5, "uom": "cm"},
            "status": "A",
        },
    ]


#Creating Database
database = client['Inventory']

#Creating Collection/Document inside Database (Same as creating Tables inside Database in SQL Systems)
collection = database["Table"]

#Storing one or multiple records/documents in collection/document
collection.insert_many(data)

#Fetching all Data
d1= collection.find()
for i in d1:
    print(i)

#Extract where status = 'A'
d2 = collection.find({'status' : 'A'})
for i in d2:
    print(i)

#Extract where status = 'A' or status = 'D'
d3 = collection.find({'status' : {'$in':['A','P']}})
for i in d3:
    print(i)

#Extract where status > 'C'
d4 = collection.find({'status' : {'$gt':'C'}})
for i in d4:
    print(i)

#Extract where quantity >= 100
d5 = collection.find({'qty' : {'$gte': 100}})
for i in d5:
    print(i)

#Extract where qty = 95 and item = 'sketch pad'
d6 = collection.find({'item' : 'sketch pad','qty' : 95})
for i in d6:
    print(i)

#Extract where qty >= 75 and item = 'sketch pad'
d7 = collection.find({'item' : 'sketch pad','qty' : {'$gte' : 75}})
for i in d7:
    print(i)

#Extract where qty >= 75 or item = 'sketch pad'
d8 = collection.find({'$or' : [{'item' : 'sketch pad'},{'qty' : {"$gte" : 75}}]})
for i in d8:
    print(i)

#'Canvas' will be updated to 'Gaurav'
collection.update_one({'item' : 'canvas'},{'$set':{'item' : 'Gaurav'}}) #Will update for one record 
d = collection.find({'item' : 'Gaurav'})
for i in d:
    print(i)

#Delete where item = 'Gaurav'
collection.delete_one({'item' : 'Gaurav'}) #Will delete for one record of item = 'Gaurav'
d = collection.find({'item' : 'Gaurav'})
for i in d:
    print(i)
```
